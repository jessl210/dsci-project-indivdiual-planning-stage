```r
### Run this cell before continuing.
library(tidyverse)
library(repr)
library(tidymodels)
options(repr.matrix.max.rows = 6)
```

```r
players <- read_csv("players.csv")
players
```

```r
sessions <- read_csv("sessions.csv")
sessions
```

```r
players_mean <- players |>
    mutate(played_hours_mean = mean(played_hours, na.rm = TRUE)) |>
    mutate(age_mean = mean(Age, na.rm = TRUE)) |>
    select(played_hours_mean, age_mean)|>
    slice(1)
players_mean
```

```r
players_experience_bar <- players |>
    ggplot(aes(x = experience, fill = as_factor(subscribe))) +
    geom_histogram(stat = "count") +
    labs(x = "Player's experience", fill = "Subcription", title = "Experience and Subscription")
players_experience_bar 
```

```r
players_gender_bar <- players |>
    ggplot(aes(x = gender, fill = as_factor(subscribe))) +
    geom_histogram(stat = "count") +
    labs(x = "Gender", fill = "Subcription", title = "Gender and Subscription")
players_gender_bar 
```

```r
players_age_hours <- players |>
    ggplot(aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point(alpha = 0.6) +
    labs(x = "Age (years)", y = "Number of hours played (hours)", color = "Subscription", title = "Hours played vs. Age and Subscription")
players_age_hours
```
